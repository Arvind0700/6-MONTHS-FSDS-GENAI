# Stacked Bar Chart

# we have to use a parameter "bottom" from plt.bar() function.

```python
a=[45,34,19,47]
b=[63,32,21,77]
c=range(4)
plt.bar(c,a,color='g')
plt.bar(c,b,color='b',bottom=a)
```

```python
a = [45, 34, 19, 47]
b = [63, 32, 21, 77]
c = range(4)

plt.bar(c, a, color='g', label='A')
plt.bar(c, b, color='b', bottom=a, label='B')

plt.title("Stacked Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
```

# Pie Chart

```python
plt.figure(figsize=(8,7))
x11=[12,16,15,30]
labels=['Pen','Paper','Eraser','Pencil']
plt.pie(x11,labels=labels)
```

```python
plt.figure(figsize=(8, 7))
x11 = [12, 16, 15, 30]
labels = ['Pen', 'Paper', 'Eraser', 'Pencil']

plt.pie(x11, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True)
#'%1.2f%%'  25.00%

#'%.0f%%'  25%

#'%.3f%%'  25.000%
plt.title("Stationery Distribution")
```

```python
x12 = [12, 16, 15, 30]
labels = ['Pen', 'Paper', 'Eraser', 'Pencil']

def custom_autopct(pct, allvals):
    absolute = int(round(pct/100.*sum(allvals)))
    return f"{labels[custom_autopct.index]}\n{pct:.1f}%"

custom_autopct.index = 0  # static variable for tracking

def wrapper(pct):
    result = custom_autopct(pct, x11)
    custom_autopct.index += 1
    return result

plt.figure(figsize=(8,7))
plt.pie(x12, labels=[None]*len(labels), autopct=wrapper, startangle=90)
plt.title("Pie Chart with Label Above Percentage")
plt.show()
```

# Boxplot

```python
x13=np.random.randn(200)
plt.boxplot(x13)
# compare distribution of values by showing median,quartiles,maximum,minimum
```

The boxplot() function takes a set of values and computes the mean, median and other statistical quantities. The following points describe the preceeding boxplot:

 The red bar is the median of the distribution.

 The blue box includes 50 percent of the data from the lower quartile to the upper quartile. Thus, the box is centered on the median of the data.

 The lower whisker extends to the lowest value within 1.5 IQR from the lower quartile.

 The upper whisker extends to the highest value within 1.5 IQR from the upper quartile.

 Values further from the whiskers are shown with a cross marker.
